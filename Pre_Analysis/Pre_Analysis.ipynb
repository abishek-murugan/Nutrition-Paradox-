```python
import pycountry
```

```python
def get_country_name(alpha_3_code):
    special_cases = {
                    'GLOBAL': 'Global',
                    'WB_LMI': 'Low & Middle Income',
                    'WB_HI': 'High Income',
                    'WB_LI': 'Low Income',
                    'EMR': 'Eastern Mediterranean Region',
                    'EUR': 'Europe',
                    'AFR': 'Africa',
                    'SEAR': 'South-East Asia Region',
                    'WPR': 'Western Pacific Region',
                    'AMR': 'Americas Region',
                    'WB_UMI': 'Upper Middle Income'}
    if alpha_3_code in special_cases.keys():
            return special_cases[alpha_3_code]
    try:
        country = pycountry.countries.get(alpha_3 = alpha_3_code)
        if country:
            return country.name
    except (AttributeError, LookupError):
        pass
    
    return alpha_3_code
```

```python
df_obesity["Country"] = df_obesity["Country"].apply(get_country_name)
```

```python
df_malnutrition["Country"] = df_malnutrition["Country"].apply(get_country_name)
```

# Data Visuals

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
region_avg = (
    df_obesity
    .groupby("Region", as_index=False)["Mean_Estimate"]
    .mean()
    .sort_values("Mean_Estimate", ascending=False)
)

fig = px.bar(
    region_avg,
    x="Region",
    y="Mean_Estimate",
    title="Average Obesity by Region",
    color="Region",
    color_continuous_scale="Reds"
)

fig.update_layout(
    yaxis_title="Average Obesity (%)"
)
```

```python
fig = px.box(
    df_obesity,
    x="Age_Group",
    y="Mean_Estimate",
    color="Age_Group",
    title="Adult vs Child Obesity Distribution"
)

fig.update_layout(
    xaxis_title="Age Group",
    yaxis_title="Obesity (%)"
)
```

```python
gender_avg = (
    df_obesity
    .groupby("Gender", as_index=False)["Mean_Estimate"]
    .mean()
)

fig = px.bar(
    gender_avg,
    x="Gender",
    y="Mean_Estimate",
    title="Average Obesity by Gender",
    color="Gender"
)

fig.update_layout(
    yaxis_title="Average Obesity (%)"
)
```
